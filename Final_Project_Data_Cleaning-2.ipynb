```python
# Convert to pandas df
import os
import pandas as pd
import json

final_df = pd.DataFrame()
for file in os.listdir("/content/drive/My Drive/data"):
  if file.endswith(".json"):
    tracklists_data = json.load(open("/content/drive/My Drive/data/" + file))
    # Initialize data
    songs_df = pd.json_normalize(tracklists_data)
    songs_df = songs_df.explode('playlists', ignore_index = False)
    # Create playlist-level df
    playlist_df = pd.json_normalize(songs_df['playlists'])
    # reset index
    playlist_df.reset_index(inplace=True)
    tracks = playlist_df['tracks'].explode()
    total_df = playlist_df.drop(columns = 'tracks').join(pd.DataFrame(tracks.tolist(), index = tracks.index).rename(columns={'duration_ms':'track_duration_ms'}))

    final_df = final_df.append(total_df)
final_df['track_uri'] = final_df['track_uri'].str.replace('spotify:track:','')
```

```python
track_uri = list(set(list(final_df['track_uri'])))
track_dict = {}
i = 0
for track in track_uri:
  track_dict[track] = i
  i = i + 1
```

```python
def find(track_uri):
  return track_dict[track_uri]
```

```python
# Create songs df
import pandas as pd

song_attributes_df = pd.read_csv('/content/drive/MyDrive/tracks_features.csv')
song_attributes_df
```

```python
# Create df that matches playlist to song
contains_df = final_df[['pid', 'track_uri','pos']]
contains_df['track_uri'] = contains_df['track_uri'].apply(lambda x: find(x))

# Create df for all song information
song_df = final_df[['artist_name', "track_uri","artist_uri","track_name","album_uri","duration_ms","album_name"]].drop_duplicates()
song_df = song_df.drop('duration_ms', axis =1)
song_df = song_df.merge(song_attributes_df, left_on = 'track_uri', right_on = 'id', how = 'left')
song_df = song_df.drop(['id','key', 'album'], axis =1)
song_df['track_uri'] = song_df['track_uri'].drop_duplicates()
song_df = song_df.dropna(subset = 'track_uri')
song_df['track_uri'] = song_df['track_uri'].apply(lambda x: find(x))

# Create df for all playlist information
playlist_df = final_df[["name","collaborative","pid", "modified_at","num_albums","num_tracks", "num_followers","num_edits","duration_ms", "num_artists"]].drop_duplicates()
```
